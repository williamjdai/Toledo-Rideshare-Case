```python
import math
```

```python
import matplotlib.pyplot as plt
```

```python
takes = [i for i in range(0,601,50)] #the company's take per ride expressed in cents
```

```python
revenues = [12*100*t for t in takes] #revenue over a year, expressed in cents, for each corresponding take
```

```python
driver_cac = 12*50000*0.20 #cost of Driver CAC over a year, expressed in cents
```

```python
rider_cac = [12*100*(min(math.log(100)+1000, 2000)*0.10*(0.60+0.11*(600-t)/100)+min(math.log(100)+1000, 2000)*0.33*(1-(0.60+0.11*(600-t)/100))) for t in takes] #cost of rider CAC over a year, expressed in cents, for each corresponding take
```

```python
profits = [revenues[i]-driver_cac-rider_cac[i] for i in range(len(takes))]
```

```python
plt.figure(figsize = (10,10))
plt.scatter(takes, profits)
plt.xlabel("Ride-share's take (cents)")
plt.ylabel("Annual profit (cents)")
plt.title("Ride-share profits using logarithmic rider acquisition cost model")
```

```python
max_profit = -99999999
for i in range(len(profits)):
    if profits[i]>max_profit:
        max_profit = profits[i]
        optimal_take = takes[i]
print(optimal_take, max_profit)
```

```python
rider_cac_alt = [12*100*(1000*0.10*(0.60+0.11*(600-t)/100)+1000*0.33*(1-(0.60+0.11*(600-t)/100))) for t in takes] #cost of rider CAC over a year, expressed in cents, for each corresponding take
```

```python
profits_alt = [revenues[i]-driver_cac-rider_cac_alt[i] for i in range(len(takes))]
```

```python
plt.figure(figsize = (10,10))
plt.scatter(takes, profits_alt)
plt.xlabel("Ride-share's take (cents)")
plt.ylabel("Annual profit (cents)")
plt.title("Ride-share profits using flat $10 rider acquisition model")
```

```python
max_profit = -99999999
for i in range(len(profits_alt)):
    if profits_alt[i]>max_profit:
        max_profit = profits_alt[i]
        optimal_take = takes[i]
print(optimal_take, max_profit)
```

```python
rider_cac_alt2 = [12*100*(2000*0.10*(0.60+0.11*(600-t)/100)+2000*0.33*(1-(0.60+0.11*(600-t)/100))) for t in takes] #cost of rider CAC over a year, expressed in cents, for each corresponding take
```

```python
profits_alt2 = [revenues[i]-driver_cac-rider_cac_alt2[i] for i in range(len(takes))]
```

```python
plt.figure(figsize = (10,10))
plt.scatter(takes, profits_alt2)
plt.xlabel("Ride-share's take (cents)")
plt.ylabel("Annual profit (cents)")
plt.title("Ride-share profits using flat $20 rider acquisition model")
```
